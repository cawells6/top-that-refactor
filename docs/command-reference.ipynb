# Command Reference

Short notebook documenting the development run commands and the fixes applied to stop automatic browser opens and to repair `~/.bashrc`.


## Key NPM scripts

- `npm run run:noopen` — starts the workspace development environment without auto-opening a browser (sets `NO_AUTO_OPEN=1`).
- `npm run dev:client` — client dev server (project configured to avoid `--open`).
- `run` — legacy launcher script (workspace run wraps the dev start).


````bash
# Start the dev environment without opening a browser
npm run run:noopen
````

## run.js behaviour

- `run.js` watches child process output and will open the browser exactly once when the Vite client URL appears, unless `NO_AUTO_OPEN` is set.
- To force no auto-open, set `NO_AUTO_OPEN=1` (the `run:noopen` script does this for you).


## Restoring ~/.bashrc (backup)

A backup was created at `~/.bashrc.bak` when the broken profile was fixed. To restore the backup and re-source your shell:

```bash
cp ~/.bashrc.bak ~/.bashrc
source ~/.bashrc
```


````bash
# Example: run the workspace task (from shell)
# Using the workspace quick no-open script:
./run
# or directly:
npm run run:noopen
````

## Notes & Troubleshooting

- If you still see duplicate browser tabs, check for user-level Action Buttons or extensions that might call `start firefox ... && run` outside the workspace settings.
- After workspace edits to `settings.json` or `tasks.json`, reload VS Code so the Action Buttons UI reflects the updated buttons.
